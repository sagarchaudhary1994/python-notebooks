Question1
Create a function that takes three integer arguments (a, b, c) and returns the amount of integers which are of equal value.
Examples
equal(3, 4, 3) ➞ 2

equal(1, 1, 1) ➞ 3

equal(3, 4, 1) ➞ 0 
Notes
Your function must return 0, 2 or 3.


```python
def equal(a,b,c):
    try:
        if type(a)!=int or type(b)!=int or type(c)!=int:
            raise Exception("Please enter integer in a, b,c")
        num_list=list((a,b,c))
        unique_num=set(num_list)
        len_diff=len(num_list)- len(unique_num)
        equal=len_diff+1 if len_diff>0 else len_diff
        return equal
    except  Exception as e:
        print("Error occurred-",e)
    
##function call
print(equal(3, 4, 3)) #expected 2
print(equal(1, 1, 1)) #expected 3
print(equal(3, 4, 1)) #expected 0
```

Question2
Write a function that converts a dictionary into a list of keys-values tuples.
Examples
dict_to_list({
  "D": 1,
  "B": 2,
  "C": 3
}) ➞ [("B", 2), ("C", 3), ("D", 1)]

dict_to_list({
  "likes": 2,
  "dislikes": 3,
  "followers": 10
}) ➞ [("dislikes", 3), ("followers", 10), ("likes", 2)]
Notes
Return the elements in the list in alphabetical order.


```python
def dict_to_list(_dict):
    try:
        if type(_dict)!=dict:
            raise TypeError("Please enter dictionary only")
        if len(_dict)<1:
            return
        items_sorted = sorted(_dict.items(),key= lambda kv:(kv[0],kv[1]))
        return items_sorted
    except Exception as e:
        print("Error encountered- ",e)
        
#function call
print(dict_to_list({ "D": 1, "B": 2, "C": 3 })) #expected- [("B", 2), ("C", 3), ("D", 1)]
print(dict_to_list({ "likes": 2, "dislikes": 3, "followers": 10 }))#expected- [("dislikes", 3), ("followers", 10), ("likes", 2)] 
```

Question3
Write a function that creates a dictionary with each (key, value) pair being the (lower case, upper case) versions of a letter, respectively.
Examples
mapping(["p", "s"]) ➞ { "p": "P", "s": "S" }

mapping(["a", "b", "c"]) ➞ { "a": "A", "b": "B", "c": "C" }

mapping(["a", "v", "y", "z"]) ➞ { "a": "A", "v": "V", "y": "Y", "z": "Z" }
Notes
All of the letters in the input list will always be lowercase.


```python
class pass_lower_case_characters(Exception):
    msg="Please pass only lower case characters"
    def __init__(self):
        Exception.__init__(self,self.msg)
    

def mapping(char_lst):
    try:
        if len(char_lst)<1:
            raise Exception("Please pass list with atleast one item!!")
        temp_dict={}
        for i in char_lst:
            if type(i)!=str:
                raise pass_lower_case_characters
            if len(i)!=1 and (ord(i)>=ord("a") or (ord(i)<=ord("z"))):
                raise pass_lower_case_characters
            if i not in temp_dict:
                temp_dict[i]=chr(ord(i)-32)
            else:
                continue
        return dict(sorted(temp_dict.items(),key= lambda kv:(kv[0],kv[1])))
        
    except Exception as e:
        print("Error encountered-",e)

#function call
print(mapping(["p", "s"]))#expected { "p": "P", "s": "S" }
print(mapping(["a", "b", "c"])) #expected { "a": "A", "b": "B", "c": "C" }
print(mapping(["a", "v", "y", "z"])) #expected { "a": "A", "v": "V", "y": "Y", "z": "Z" }
```

Question4
Write a function, that replaces all vowels in a string with a specified vowel.
Examples
vow_replace("apples and bananas", "u") ➞ "upplus und bununus"

vow_replace("cheese casserole", "o") ➞ "chooso cossorolo"

vow_replace("stuffed jalapeno poppers", "e") ➞ "steffed jelepene peppers"
Notes
All words will be lowercase. Y is not considered a vowel.


```python
class all_lowercase(Exception):
    msg="Please input string with all lower case character"
    def __init__(self):
        Exception.__init__(self,self.msg)
        
def vow_replace(string,replace_with):
    try:
        vowel=["a","e","i","o","u"]
        if type(string)!=str:
            raise all_lowercase
        if type(replace_with)!=str:
            raise TypeError("Please enter vowel to replace other vowels")
        if replace_with not in vowel:
            return
        lst_words=list(string)
        for i in range(len(lst_words)):
            char=lst_words[i]
            if char in vowel:
                lst_words[i]=replace_with
        return "".join(lst_words)
    except Exception as e:
        print("Error encounterd-",e)
    
#function call
print(vow_replace("apples and bananas", "u")) #expected- "upplus und bununus"
print(vow_replace("cheese casserole", "o")) #expected- "chooso cossorolo"
print(vow_replace("stuffed jalapeno poppers", "e")) #expected- "steffed jelepene peppers"
```

Question5
Create a function that takes a string as input and capitalizes a letter if its ASCII code is even and returns its lower case version if its ASCII code is odd.
Examples
ascii_capitalize("to be or not to be!") ➞ "To Be oR NoT To Be!"

ascii_capitalize("THE LITTLE MERMAID") ➞ "THe LiTTLe meRmaiD"

ascii_capitalize("Oh what a beautiful morning.") ➞ "oH wHaT a BeauTiFuL moRNiNg."


```python
def ascii_capitalize(string):
    try:
        if type(string)!=str:
            raise TypeError("Please enter only string")
        lst_string=list(string)
        
        for i in range(len(lst_string)):
            char=lst_string[i]
            if (ord(char)>=ord("a") and ord(char)<=ord("z")):
                lst_string[i]=chr(ord(char)-32) if ord(char)%2==0 else char
            elif  (ord(char)>=ord("A") and ord(char)<=ord("Z")):
                lst_string[i]=chr(ord(char)+32) if ord(char)%2!=0 else char
            else:
                continue
        return "".join(lst_string)
    except Exception as e:
        print("Error encountered-",e)
        
#function call
print(ascii_capitalize("Oh what a beautiful morning.")) # expected- "oH wHaT a BeauTiFuL moRNiNg."
print(ascii_capitalize("to be or not to be!")) #expected- "To Be oR NoT To Be!"
print(ascii_capitalize("THE LITTLE MERMAID")) #expected- "THe LiTTLe meRmaiD"
```
